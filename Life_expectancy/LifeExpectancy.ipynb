7. Drop the column ‘country’ and ’status’ from the dataset and split the dataset into
training and testing in a 80:20 split. Print the head for the training set. (FYI: Target Variable is Life Expectancy) 

```python
df_dropped = df.drop(columns=['Country' , 'Status'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_dropped, df_dropped, test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
# Fit the scaler to the training data and transform
X_train_scaled = scaler.fit_transform(X_train)

# Transform the test data using the fitted scaler
X_test_scaled = scaler.transform(X_test)
```

```python

print (f"Training_data ratio : {(X_train_scaled.shape[0]/df_dropped.shape[0])*100 :.2f} : {( X_test_scaled.shape[0]/df_dropped.shape[0])*100:.2f}")
print("Scaled Training Data shape:")
print(X_train_scaled.shape)
print("Scaled Testing Data shape:")
print(X_test_scaled.shape)
print("Data shape:")
print(df_dropped.shape)
```

8. Build a linear regression model using the training and testing datasets and compute
mean absolute error. (Use sklearn) 

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
y_pred = regression.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae:.4f}")
```
